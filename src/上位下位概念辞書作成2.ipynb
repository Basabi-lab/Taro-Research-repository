```python
import csv
```

```python
from util.difficultyCheck import comparisonCheck, eraseVersion
```

```python
with open("./qiita_tags.csv", "r") as f:
    tags = list(set([i for i in map(eraseVersion, f.read().split(","))][:500]))[1:]
print(tags)
```

```python
len(tags)
```

```python
use_tags = []
results = []
start = 0
```

```python
# 開始場所
s = start
for i, tag in enumerate(tags[start:len(tags)]):
    print("["+str(s+i)+"]:", tag)
    w1 = "javascript"
    w2 = tag
    
    try:
        res = comparisonCheck(w1, w2)
        start += 1
    except:
        break
    
    if res:
        results.append(str(res))
        use_tags.append(w2.lower())
    print()
    
with open("html_easy_hard.csv", "w") as f:
    f.write(",".join(results))
```

```python
res = [eval(r) for r in results]
sorted(res, key=lambda x:x['score'], reverse=True)
```

```python
for r in sorted(res, key=lambda x:x['score'], reverse=True):
    # 開始場所
    start = 0
    s = start
    for i, tag in enumerate(tags[start:len(tags)]):
        print("["+str(s+i)+"]:", tag)
        w1 = r["hard"]
        w2 = tag

        try:
            res = comparisonCheck(w1, w2)
            start += 1
        except:
            break

        if res:
            results.append(str(res))
        print()

    with open("js_easy_hard.csv", "w") as f:
        f.write("\n".join(results))
    print(res)
    print(start)
```

```python
ts = []
for r in results:
    res = eval(r)
    if res["hard"] in use_tags:
        ts.append(res["hard"])
```

```python
# 開始場所
for r in list(set(ts)):
    start = 0
    s = start
    for i, tag in enumerate(tags[start:len(tags)]):
        print("["+str(s+i)+"]:", tag)
        w1 = r
        w2 = tag

        try:
            res = comparisonCheck(w1, w2)
            start += 1
        except:
            break

        if res:
            results.append(str(res))
            use_tags.append(w2.lower())
        print()

    with open("html_easy_hard.csv", "w") as f:
        f.write(",".join(results))
```

```python
res = results
```

```python
new_res = []

for r in res:
    flag = 0
    er = eval(r)
    for nr in new_res:
        if er["easy"].lower() == nr["easy"].lower() and er["hard"].lower() == nr["hard"].lower():
            flag = 1
            break
            
    if flag == 0:
        new_res.append({'easy': er["easy"].lower(), 'hard': er["hard"].lower(), 'score': er["score"]})
```
